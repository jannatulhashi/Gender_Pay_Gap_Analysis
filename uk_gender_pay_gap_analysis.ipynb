```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# Load in file
# Store filepath in a variable

pay_gap_data_2018_2019_csv = Path("Resources/UK Gender Pay Gap Data - 2018 to 2019.csv")
pay_gap_data_2019_2020_csv = Path("Resources/UK Gender Pay Gap Data - 2019 to 2020.csv")
pay_gap_data_2020_2021_csv = Path("Resources/UK Gender Pay Gap Data - 2020 to 2021.csv")
pay_gap_data_2021_2022_csv = Path("Resources/UK Gender Pay Gap Data - 2021 to 2022.csv")
pay_gap_data_2022_2023_csv = Path("Resources/UK Gender Pay Gap Data - 2022 to 2023.csv")
```

```python
# Read titles and credits data file and store into Pandas DataFrames

pay_gap_data_2018_2019_df = pd.read_csv(pay_gap_data_2018_2019_csv)
pay_gap_data_2019_2020_df = pd.read_csv(pay_gap_data_2019_2020_csv)
pay_gap_data_2020_2021_df = pd.read_csv(pay_gap_data_2020_2021_csv)
pay_gap_data_2021_2022_df = pd.read_csv(pay_gap_data_2021_2022_csv)
pay_gap_data_2022_2023_df = pd.read_csv(pay_gap_data_2022_2023_csv)
```

```python
# Concatenate the pay gap dataframes from the years 2018 to 2023 into a single dataframe 'pay_gap_data_df'.

pay_gap_data_df = pd.concat([pay_gap_data_2018_2019_df, pay_gap_data_2019_2020_df, pay_gap_data_2020_2021_df,
                            pay_gap_data_2021_2022_df, pay_gap_data_2022_2023_df])

# Display the DataFrame
pay_gap_data_df
```

```python
pay_gap_data_df.columns
```

# Cleaning the Dataset

```python
#count the number of missing values in each column of pay_gap_data_df

pay_gap_data_df.isnull().sum()
```

```python
# Drop EmployerId  and PostCode columns because they have a lot of  missing values

pay_gap_data_df.drop(["EmployerId", "PostCode"], axis=1, inplace=True)
```

```python
# Impute missing values with "Unknown" for the 'Address', 'CompanyNumber', 'SicCodes' columns

pay_gap_data_df["Address"].fillna("Unknown", inplace=True)
pay_gap_data_df["CompanyNumber"].fillna("Unknown", inplace=True)
pay_gap_data_df["SicCodes"].fillna("Unknown", inplace=True)
```

```python
# Generate summary statistics for the dataset

pay_gap_data_df.describe()
```

```python
# Handle missing values in gender pay gap percentage columns

pay_gap_data_df["DiffMeanHourlyPercent"].fillna(pay_gap_data_df["DiffMeanHourlyPercent"].median(), inplace=True)
pay_gap_data_df["DiffMedianHourlyPercent"].fillna(pay_gap_data_df["DiffMedianHourlyPercent"].median(), inplace=True)
pay_gap_data_df["DiffMeanBonusPercent"].fillna(pay_gap_data_df["DiffMeanBonusPercent"].median(), inplace=True)
pay_gap_data_df["DiffMedianBonusPercent"].fillna(pay_gap_data_df["DiffMedianBonusPercent"].median(), inplace=True)

# Handle missing values in quartile columns

quartile_columns = [
    "MaleLowerQuartile", "FemaleLowerQuartile", "MaleLowerMiddleQuartile", "FemaleLowerMiddleQuartile",
    "MaleUpperMiddleQuartile", "FemaleUpperMiddleQuartile", "MaleTopQuartile", "FemaleTopQuartile"]

for col in quartile_columns:
    pay_gap_data_df[col].fillna(pay_gap_data_df[col].median(), inplace=True)

# Handle missing values in text/categorical columns

pay_gap_data_df["CompanyLinkToGPGInfo"].fillna("Unknown", inplace=True)
pay_gap_data_df["ResponsiblePerson"].fillna("Unknown", inplace=True)
```

```python
# Display data types of each column in the DataFrame
pay_gap_data_df.dtypes
```

```python
# Convert the specified columns to their correct data types
pay_gap_data_df["CompanyNumber"] = pay_gap_data_df["CompanyNumber"].astype("str")
pay_gap_data_df["SicCodes"] = pay_gap_data_df["SicCodes"].astype("str")
pay_gap_data_df["EmployerSize"] = pay_gap_data_df["EmployerSize"].astype("str")
pay_gap_data_df["DueDate"] = pd.to_datetime(pay_gap_data_df["DueDate"])

```

```python
# Convert the 'DateSubmitted' column to datetime format
pay_gap_data_df["DateSubmitted"] = pd.to_datetime(pay_gap_data_df["DateSubmitted"], errors='coerce')

# Normalize the datetime to keep only the date part
pay_gap_data_df["DateSubmitted"] = pay_gap_data_df["DateSubmitted"].dt.normalize()
```

```python
# Define a dictionary to map old column names to new ones

column_mapping = {'EmployerName': 'Employer_Name',
                'Address': 'Employer_Address',
                'CompanyNumber': 'Company_Number',
                'SicCodes': 'SIC_Codes',
                'DiffMeanHourlyPercent': 'Mean_Hourly_Pay_Gap',
                'DiffMedianHourlyPercent': 'Median_Hourly_Pay_Gap',
                'DiffMeanBonusPercent': 'Mean_Bonus_Pay_Gap',
                'DiffMedianBonusPercent': 'Median_Bonus_Pay_Gap',
                'MaleBonusPercent': 'Male_Bonus_Percentage',
                'FemaleBonusPercent': 'Female_Bonus_Percentage',
                'MaleLowerQuartile': 'Male_Lower_Quartile',
                'FemaleLowerQuartile': 'Female_Lower_Quartile',
                'MaleLowerMiddleQuartile': 'Male_Lower_Middle_Quartile',
                'FemaleLowerMiddleQuartile': 'Female_Lower_Middle_Quartile',
                'MaleUpperMiddleQuartile': 'Male_Upper_Middle_Quartile',
                'FemaleUpperMiddleQuartile': 'Female_Upper_Middle_Quartile',
                'MaleTopQuartile': 'Male_Top_Quartile',
                'FemaleTopQuartile': 'Female_Top_Quartile',
                'CompanyLinkToGPGInfo': 'Link_to_GPG_Information',
                'ResponsiblePerson': 'Responsible_Person',
                'EmployerSize': 'Employer_Size',
                'CurrentName': 'Current_Name',
                'SubmittedAfterTheDeadline': 'Submitted_After_the_Deadline',
                'DueDate': 'Due_Date',
                'DateSubmitted': 'Date_Submitted'
}

# Rename the columns
pay_gap_data_df.rename(columns=column_mapping, inplace=True)
```

# Analysis of the Data

### **How have the Mean and Median Hourly Pay Gaps changed over the years, and what does the trend indicate about gender pay equity?**

```python
# Extracting the reporting year from the 'Due_Date' column

pay_gap_data_df['Year'] = pd.to_datetime(pay_gap_data_df['Due_Date']).dt.year

# Group by Year and compute the mean for the required columns
yearly_gaps = pay_gap_data_df.groupby('Year')[['Mean_Hourly_Pay_Gap', 'Median_Hourly_Pay_Gap']].mean()

# Plotting the data
plt.figure(figsize=(8, 5))
yearly_gaps['Mean_Hourly_Pay_Gap'].plot(label='Mean Hourly Pay Gap', marker='o')
yearly_gaps['Median_Hourly_Pay_Gap'].plot(label='Median Hourly Pay Gap', marker='o')
plt.title('Hourly Pay Disparity Over Time')
plt.xlabel('Year')
plt.ylabel('Pay Gap Percentage')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/LineChart.png")

plt.show()
```

**Analysis:**

The graph showcases the evolution of the Mean and Median Hourly Pay Gaps from 2019 to 2023. From 2019 to 2021, the Mean Hourly Pay Gap decreased, indicating progress in addressing gender pay disparity. However, the Median Hourly Pay Gap showed a sharp increase during the same period but then saw a significant decline by 2022. By 2023, both gaps have reduced, suggesting improvements in gender pay equity. It's essential to note that while there's a decrease over time, there's still a noticeable disparity, highlighting the need for continued efforts in addressing this issue.

In simpler term,  While there's been some progress in paying men and women equally over the years, there's still a difference in their pay. This means more efforts are needed to close this pay gap such as:

- Review hiring and promotion practices to ensure fairness.
- Offer training programs to reduce unconscious bias.
- Encourage open discussions about pay within the company.
- Regularly review and adjust salaries to make sure they're equal for similar roles.
- Promote more women into leadership positions.

Taking these steps can not only help close the pay gap but also create a more inclusive and fair workplace for everyone. It's essential for businesses to be proactive in addressing these disparities to ensure equal opportunities for all employees.


### **How do the average bonus percentages differ between males and females?**

```python
# Calculate the average bonus percentage for males and females
male_bonus_avg = pay_gap_data_df['Male_Bonus_Percentage'].mean()
female_bonus_avg = pay_gap_data_df['Female_Bonus_Percentage'].mean()

# Create the bar chart
labels = ['Male', 'Female'] # Define the labels
values = [male_bonus_avg, female_bonus_avg] # Create a list of the average bonus percentages we computed for each gender.
colors = ['blue', 'pink']
width = 0.5
bars = plt.bar(labels, values, color=colors, width=width)
plt.xlabel('Gender')
plt.ylabel('Average Bonus Percentage (%)')  # Updated ylabel to indicate percentages
plt.title('Bonus Percentage Difference Between Genders')

# Annotate the bar chart with values
for bar in bars:
    yval = bar.get_height() # Get the height (value) of the bar
    
    # Place a text above each bar, at the center, showing its value
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, "{:.2f}%".format(yval), 
             ha='center', va='bottom')  # Use "{:.2f}%" to format as percentage

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/BarChart_1.png")

plt.show()
```

**Analysis:**

Based on the bar chart, males receive an average bonus percentage of 37.35%, while females receive an average bonus percentage of 36.20%. The difference between the bonus percentages received by males and females is about 1.15%, which is relatively small. This suggests that both genders receive nearly equivalent bonus percentages on average.

### **What is the average representation of male and female employees in the top wage quartile in the UK?**

```python
# Average representation in the top quartile
avg_male_top_quartile = pay_gap_data_df['Male_Top_Quartile'].mean()
avg_female_top_quartile = pay_gap_data_df['Female_Top_Quartile'].mean()

# Plot the bar chart
labels = ['Male', 'Female']
values = [avg_male_top_quartile, avg_female_top_quartile]
colors = ['orange', 'purple']
width = 0.5
bars = plt.bar(labels, values, color=colors, width=width)

# Annotating bars with their respective values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1,  # Add 1 to the yval for a little space above the bar
             '{:.2f}%'.format(yval),
             ha='center', va='bottom', fontsize=10)

plt.xlabel('Gender')
plt.ylabel('Average Representation in Top Quartile (%)')
plt.title('Representation in Higher Wage Quartiles')
plt.ylim(0, 105)  # Adjusted y-axis limits for space for the annotations

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/BarChart_2.png")

plt.show()
```

**Analysis:**

By analyzing this chart, we can see a difference in the number of men and women who earn higher salaries. Specifically, a bigger percentage of men (59.59%) are in the top-paying jobs compared to women (40.41%). This shows that women might not be getting as many opportunities for higher-paying positions as men. It's essential to recognize this gap and work towards giving everyone, regardless of gender, an equal chance at these top jobs.

###  **How many companies are giving bonuses to men compared to women?**

```python
# Count the number of companies that gave bonuses to males (where Male_Bonus_Percentage is greater than 0)
male_bonus_given = (pay_gap_data_df['Male_Bonus_Percentage'] > 0).sum()

# Count the number of companies that gave bonuses to females (i.e., where Female_Bonus_Percentage is greater than 0)
female_bonus_given = (pay_gap_data_df['Female_Bonus_Percentage'] > 0).sum()

# Plot a pie chart with the bonus data
bonus_data = [male_bonus_given, female_bonus_given]
labels = ['Companies giving bonuses to Males', 'Companies giving bonuses to Females']
plt.pie(bonus_data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
plt.title("Distribution of Companies Providing Bonuses by Gender")

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/PieChart.png")

plt.show()
```

**Analysis:**

The pie chart provides a clear snapshot of inclusivity in bonus distribution across companies. There's a commendable near-equal split, with 50.2% of companies awarding bonuses to males and 49.8% to females. This suggests a trend toward gender balance in recognizing employee contributions. While this is an encouraging sign consistent of promoting inclusivity, it's also crucial to delve deeper and assess whether the value and size of these bonuses are equitable across genders to ensure comprehensive fairness.

### **How many companies pay men more than women and how many pay women more than men?**

```python
# Count the number of employers where the Mean_Hourly_Pay_Gap is greater than 0, indicating men earn more on average
positive_gap = (pay_gap_data_df['Mean_Hourly_Pay_Gap'] > 0).sum()

# Count the number of employers where the Mean_Hourly_Pay_Gap is less than 0, indicating women earn more on average
negative_gap = (pay_gap_data_df['Mean_Hourly_Pay_Gap'] < 0).sum()

# Plot the bar chart
width=0.5
plt.bar(['Positive Gap', 'Negative Gap'], [positive_gap, negative_gap], color=['green', 'red'], width=width)
plt.title("Employers with Positive vs Negative Mean Hourly Pay Gap")
plt.ylabel("Number of Employers")

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/BarChart_3.png")

plt.show()
```

**Analysis:**

The bar chart reveals a significant disparity in hourly pay across companies. A vast majority of employers have a positive pay gap, meaning men generally earn more than women. In contrast, only a small fraction of companies have a negative pay gap, where women earn more than men. This prominent inequality underscores the need for greater efforts in promoting gender pay equity.  It is essential to encourage businesses to embrace, and adopt more inclusive practices, and work diligently to overcome this pay gap. This isn't just about fairness; it's about recognizing and valuing contributions irrespective of gender.

### **What are the top 10 sectors of work (based on SIC codes) where the gender hourly pay difference is the greatest?**

**Analysis:**

This chart lists the top 10 job areas( (based on SIC codes)  with big pay differences between men and women. Starting from the left, the teal bar shows the sector where men earn more than women. Across the chart, the differences in pay are substantial, indicating that in all these areas, men generally earn more than women. The goal should be achieving pay equality across all sectors, underscoring the importance of promoting inclusivity in the workplace.

### **How does the average bonus pay difference between genders vary across different employer sizes?**

```python
# Group the data by 'Employer_Size' and compute the mean for 'Mean_Bonus_Pay_Gap' for each employer size category.
bonus_by_size = pay_gap_data_df.groupby('Employer_Size')['Mean_Bonus_Pay_Gap'].mean().sort_values()

# Create a line plot
ax = bonus_by_size.plot(kind='line', marker='o', color='cyan')
plt.title("Mean Bonus Pay Gap by Employer Size")
plt.ylabel("Mean Bonus Pay Gap (%)")
plt.xticks(rotation=75)
plt.tight_layout()
plt.grid(True)

# Annotate each data point on the line plot with its value.
x_ticks = ax.get_xticks()

# Loop through each data point (x, y) on the line and annotate it with the y value
for i, (x, y) in enumerate(zip(x_ticks, bonus_by_size)):    
    plt.text(x, y + 0.005, round(y, 2), ha='center', va='bottom', fontsize=8, rotation=0)

# Save an image of the chart as a PNG file
plt.savefig("Visualizations/LineChart_2.png")

plt.show()
```

**Analysis:**

This graph illustrates how the bonus pay gap between genders changes as companies get larger. Starting with smaller firms of 250-499 employees, the difference in bonus pay between genders is around 23%. As companies grow in size, this gap tends to increase, peaking near 30% for companies with 5,000 to 19,999 employees. This suggests that larger companies, on average, have a more significant difference in bonus wages between men and women.

In general, the bigger the company, the larger the difference in bonus payments between male and female employees.
It's essential for companies, especially larger ones, to work towards ensuring bonuses are distributed equitably. For larger organizations, it's crucial to prioritize inclusivity, ensuring that bonuses are given fairly across genders.
